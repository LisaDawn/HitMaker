```python
import pandas as pd
import numpy as np
import scipy.spatial
import json
import random
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#initialize vectorizer
vectorizer = CountVectorizer(min_df=1)
vectorizer
```

```python
#read in json file
corpus = json.load(open("articles_html1000.json"))
```

```python
#prep data by labeling - helpful in plotting
section_corp = []
section_headline = []
section_label = []
joint_label = []

# for article arrays and labeling
for article in corpus:
    # some of the articles are split into multiple lists. join them
    str_article = (" ").join(article['content'])
    # make list of article texts
    section_corp.append(str_article)
    # various labels for plotting
    joint_label.append(article['section_name'] + '-' + article['headline']['main'])
    section_label.append(article['section_name'])
    section_headline.append(article['headline']['main'])
```

```python
#vectorize the matrix of text in articles
X = vectorizer.fit_transform(section_corp)
X
```
